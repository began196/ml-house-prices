```python
# Specify libraries
import pandas as pd
import numpy as np
```

```python
# Read in raw data
train_data = pd.read_csv('data/data_processed.csv')
test_data = pd.read_csv('data/data_test_processed.csv')

# Check rows and columns
train_data.head()
```

```python
from scripts.one_hot_encoder import onehotencode_data

# Specify dependent variable and independent variables
train_Y = train_data.SalePrice
train_X = train_data.drop(columns = ['Id', 'SalePrice'])
test_X = test_data.drop(columns = ['Id'])

train_X = onehotencode_data(train_X)
test_X = onehotencode_data(test_X)
```

```python
# Model number 1
X = train_X
y = train_Y

from sklearn.model_selection import train_test_split, KFold, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn import metrics

# training the model on training set
lin_reg = LinearRegression()

# Perform k-fold cross-validation
kf = KFold(n_splits=5, shuffle=True)
mse_scores = cross_val_score(lin_reg, X, y, scoring='neg_mean_squared_error', cv=kf)
r2_scores = cross_val_score(lin_reg, X, y, scoring='r2', cv=kf)

# Convert negative MSE scores to positive
mse_scores = -mse_scores

# Print cross-validation results
print(f"Cross-Validated Mean Squared Error: {np.mean(mse_scores)} Â± {np.std(mse_scores)}")
print(f"Cross-Validated R-squared: {np.mean(r2_scores)} Â± {np.std(r2_scores)}")
```
